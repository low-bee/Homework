```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_excel(r"city.xls", sheet_name=10)
```

```python
x, y = data["t"], data["y"]
```

```python
plt.figure()
plt.scatter(x,y)
plt.xlabel("T", size=18)
plt.ylabel("GDP", size=20)
plt.show()
```

```python
model = LinearRegression(fit_intercept=True)
x = np.array(x).reshape(-1,1)
model.fit(x,y)
x_prediction = np.linspace(0, 30, 23).reshape(-1, 1)
y_prediction = model.predict(x_prediction)
```

```python
plt.figure()
plt.scatter(x, y)
plt.plot(x_prediction, y_prediction, color="red")
plt.xlabel("T", size=18)
plt.ylabel("GDP", size=20)
plt.show()
```

```python
r2_score(y, y_prediction)
```

```python
ploy_model= PolynomialFeatures(degree=3)
x_ploy = ploy_model.fit_transform(x)
lin_reg = linear_model.LinearRegression()
lin_reg.fit(x_ploy,y)
```

```python
x_1_pred = lin_reg.predict(x_ploy)
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(x_prediction, y_prediction,color="red", label = "y = t0 + t1 * x")
plt.plot(x, x_1_pred, label="y = t0 + t1 * x + t2 * x^2")
plt.legend()
plt.xlabel("T",size=18)
plt.ylabel("GDP",size=20)
plt.show()
```
